```python
import random as random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 100
prices = [(0, S0)]
S1 = S0

for t in range(1, 101):
    shock = max(min(random.gauss(0, 1), 1), -1)/10 
    S2 = S1*(1+(shock))
    prices.append((t, S2))
    S1 = S2

print(prices)

df = pd.DataFrame(prices, columns=["time", "price"])

df.plot(x="time", y="price", figsize=(12,6))
plt.show()
```
