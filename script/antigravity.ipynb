```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import random as random
```

```python
St = 100
paths = [St]

for i in range(1,1001):
    St = St * random.uniform(0.999, 1.001)
    paths.append(St)

plt.figure(figsize=(12, 6))
plt.plot(paths)
plt.show()
```
