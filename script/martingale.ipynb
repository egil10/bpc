```python
import numpy as np 
import matplotlib.pyplot as plt
import random as rd
from scipy import stats

def casino(a, b, m):
    
    doublings = []

    for i in range(1,m):
        x = a
        y = b
        n = 0  # Count doublings
        
        while x > y:
            x = rd.randint(0,a)
            y = rd.randint(0,b)
            n = n + 1

        doublings.append(n)   

    return doublings

results = casino(1000, 1, 1000)

# Create density plot
density = stats.gaussian_kde(results)
xs = np.linspace(min(results), max(results), 200)

plt.figure(figsize=(12, 6))
plt.plot(xs, density(xs), linewidth=2, color='steelblue')
plt.xlabel('Number of Doublings', fontsize=12)  
plt.ylabel('Density', fontsize=12)    
plt.title('Distribution of Doublings Before Win', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.fill_between(xs, density(xs), alpha=0.4, color='lightblue')
plt.tight_layout()
plt.show()

# Print some stats
print(f"Average doublings: {np.mean(results):.2f}")
print(f"Max doublings: {max(results)}")
print(f"Min doublings: {min(results)}")
```
