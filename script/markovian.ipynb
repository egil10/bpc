```python
import numpy as np
import matplotlib.pyplot as plt
from poisson_price_process import poisson_price_paths
```

```python
# poisson_price_paths(price0=100, lam=5, T=10, sigma=0.01)
markov = poisson_price_paths(price0 = 100, lam=10, T=10000, sigma=0.01)

plt.figure(figsize=(18,6))
plt.step(markov[:,0], markov[:,1], where='post', lw=1.5)
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price Path with Poisson Trade Arrivals")
plt.grid(True)
plt.tight_layout()
plt.savefig("plots/poisson_price_path.pdf", format="pdf")
plt.show()
plt.close()
```

```python
# poisson_price_paths(price0=100, lam=5, T=10, sigma=0.01)
markov = poisson_price_paths(price0 = 100, lam=100, T=10, sigma=0.1)

plt.figure(figsize=(18,6))
plt.step(markov[:,0], markov[:,1], where='post', lw=1.5)
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price Path with Poisson Trade Arrivals")
plt.grid(True)
plt.tight_layout()
plt.savefig("plots/poisson_price_path_t10.pdf", format="pdf")
plt.show()
plt.close()
```

```python
markov = poisson_price_paths(price0=100, lam=10, T=10000, sigma=0.01)

# --- Compute returns ---
prices = markov[:, 1]
log_returns = np.diff(np.log(prices))  # log returns between trades

# --- Plot histogram ---
plt.figure(figsize=(18,6))
plt.hist(log_returns, bins=250, density=True, alpha=0.7, color='blue')
plt.xlabel("Log Return")
plt.ylabel("Density")
plt.title("Distribution of Log Returns (Poisson Price Process)")
plt.grid(True)
plt.tight_layout()
plt.savefig("plots/poisson_price_returns_hist.pdf", format="pdf")
plt.show()
plt.close()
```

```python
plt.figure(figsize=(18,6))
plt.boxplot(log_returns, vert=True, patch_artist=True)
plt.title("Boxplot of Log Returns")
plt.ylabel("Log Return")
plt.grid(True)
plt.tight_layout()
plt.savefig("plots/poisson_returns_boxplot.pdf", format="pdf")
plt.show()
plt.close()
```

```python
from statsmodels.distributions.empirical_distribution import ECDF

ecdf = ECDF(log_returns)
plt.figure(figsize=(8,5))
plt.plot(ecdf.x, ecdf.y, lw=1.5)
plt.title("Empirical CDF of Log Returns")
plt.xlabel("Log Return")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.tight_layout()
plt.savefig("plots/poisson_returns_ecdf.pdf", format="pdf")
plt.show()
plt.close()
```
