```python
import numpy as np
import matplotlib.pyplot as plt
from poisson_price_process import poisson_price_paths
```

```python
plt.figure(figsize=(18,6))

for i in range(1, 6):  # five paths
    markov = poisson_price_paths(price0=100, lam=10, T=500, sigma=0.01)
    plt.step(markov[:,0], markov[:,1], where='post', lw=1, alpha=0.8, label=f'Path {i}')

plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Five Simulated Price Paths with Poisson Trade Arrivals")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("plots/poisson_price_paths_5.pdf", format="pdf")
plt.show()
plt.close()
```
