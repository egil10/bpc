## **up-down**

```python
import random as random
import matplotlib.pyplot as plt

S0 = 0
path = [S0]

for i in range(1, 100001):
    x = random.choice([-1, 1])
    S0 = S0+x
    path.append(S0)

print(path[:10])

plt.figure(figsize=(12,6))
plt.plot(path)
plt.show()
```
