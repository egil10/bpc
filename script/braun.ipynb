## Braun 

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import os

antallsteg = 100000
antall_iterasjoner = 100

retninger = ["hyre", "opp", "ned", "venstre"]

# Make sure plots/ exists
os.makedirs("plots", exist_ok=True)

cmap = plt.cm.rainbow  # rainbow colours

plt.figure(figsize=(10, 10))

for k in range(antall_iterasjoner):

    x = [0]
    y = [0]

    for i in range(antallsteg):
        bevegelse = random.choice(retninger)

        if bevegelse == "ned":
            y.append(y[-1] - 1)
            x.append(x[-1])

        elif bevegelse == "opp":
            y.append(y[-1] + 1)
            x.append(x[-1])

        elif bevegelse == "hyre":
            y.append(y[-1])
            x.append(x[-1] + 1)

        elif bevegelse == "venstre":
            y.append(y[-1])
            x.append(x[-1] - 1)

    # Rainbow gradient color (01)
    color = cmap(k / (antall_iterasjoner - 1))

    plt.plot(x, y, linewidth=0.8, color=color)

plt.title("50 Random Walks (Rainbow Gradient)", fontsize=18)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Save to PDF
plt.savefig("plots/random_walks.pdf", bbox_inches="tight")

plt.show()

print("Saved PDF to plots/random_walks.pdf")
```
